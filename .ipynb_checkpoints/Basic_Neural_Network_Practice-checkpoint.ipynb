# Background Research

## 19.1.1 Neural Network (Artificial Neural Networks, ANN)

Key Points: 

1. ANNs are a set of algorithms that are modeled after the human brain; they are an advanced form of machine learning that recognizes patterns and features in input data and provides a clear quantitative output.


2. The simplest form of a neural network is the layer of neurons - each of these neurons perform individual computations.


3. The computations are connected and weighted against one another, until the neurons reach the final layer, which returns a numerical result or an encoded categorical result. 


4. [`Module Question`] Difference between a numerical and categorical result: a numerical result is a predicted measurement (int, number value); a categorical result is a label (words) - it predicts which group the input will come from. 

### Ways to use a neural network model

1. Create a claffication algorithm - determines if an input belongs in one category vs another. Behaves similarly to a regression model where the dependent output variable (`target`) predicted independent input variables (`features`).


2. Can be an alternative to other models - ex: random forest, logistic regression, or linear regression models. 

### Advantages of neural network model:

1. More effective at detecting complex, nonlinear relationships.


2. Have greater tolerance for messy data and can learn to ignore noisy characteristics in data.


3. Easy to implement, scalable, and effective.

### Disadvantages of neural network models:

1. Layers of neurons are often too complex to dissect and understand (creating a black box problem)


2. Neural networks are prone to overfitting (characterizing the training data so well that it does not effectively generalize to test data.)

### Real life uses of neural networks:
   1) Finance: used to detect fraud and trends in the stock market; 
    
   2) Company profit via marketing: Amazon and Apple use neural networks to classify their consumers to provide targeted marketing as well as behavior training for robotic control systems. 

## 19.1.2 Perceptron, the Computational Neuron

Key points:

1. Computational neurons come from the field of artificial intelligence of the 1950s - the peceptron, was the machine for training the first neural network.


2. The peceptron model is a single neural network unit; it mimics a biological neuron by receiving input data, weighing the information, and producing a clear output.


##### 3. IMPORTANT: As more multidimensional perceptrons are meshed together and layered, a new, more powerful classification and regression algorithm emerges—the neural network.


#### 3. Four major components of a perceptron model:
    
    A. Input values (labelled x or X (chi))
    
    B. Weight coefficient for each input value w or w (omega)
    
    C. Bias - a constant value added to the input to influence the final decision, labelled as w0. 
    
    D. Net summary function - aggregates all weighted inputs, as a weighted summation. 



#### 4. Most common use of a perceptron model:  

1. Perceptrons are most commonly usd to separate data into two groups (also known as `linear binary classifier`) - i.e. algorithm works to classify two groups that can be separated by a linear equation (`linearly separable`). Listed below: left image (linearly separable), right group (not):



2. The perceptron model with labeled input data represents supervised machine learning because the model is provided with input and out information.


3. The perceptron model is designed to produce a discrete classification model and to learn from the input data to improve classifications as more data is analyzed. 


### How a perceptron model trains:

* The process of perceptron model training continues again and again until one of the three conditions are met:
    
    
    1. The perceptron model exceeds a predetermined performance threshold, determined by the designer before training. In machine learning this is quantified by minimizing the `loss` metric (however many epochs to have the lowest training loss score possible).


    2. The perceptron model training performs a set number of iterations, determined by the designer before training.
    
    
    3. The perceptron model is stopped or encounters an error during training.

### 19.1.3 Make the Connections and Explore TensorFlow Playground

* Now that we understand the structure of a `single neuron (i.e. perceptron)`, we can begin to build the structure of a neural network.

### Basic neural Network


* A basic neural network has three layers:

    * An input layer of input values transformed by weight coefficients
    
    * A single “hidden” layer of neurons (single neuron or multiple neurons)
    
    * An output layer that reports the classification or regression model value

* [RECALL]: Neural networks work by linking together neurons and producing a clear quantitative output. To combine each neuron with it's own output into a single classifier or regression model, and `activation function` is needed.

##### `Activation function`: mathematical function applied to the end of each "neuron" that transforms the output to a quantitative value (i.e. similar to transformation functions in supervised/unsupervised that yielded means and std dev values after scaling).


### Types of `Activation Functions`:

1. The `linear function` returns the sum of our weighted inputs without transformation.


2. The `sigmoid function` is identified by a characteristic S curve. It transforms the output to a range between 0 and 1.


3. The `tanh function` is also identified by a characteristic S curve; however, it transforms the output to a range between -1 and 1.


4. The `Rectified Linear Unit (ReLU) function` returns a value from 0 to infinity, so any negative input through the activation function is 0. It is the most used activation function in neural networks due to its simplifying output, but it might not be appropriate for simpler models.


5. The `Leaky ReLU function` is a “leaky” alternative to the ReLU function, whereby negative input values will return very small negative values.

### 19.2.2 Build a Basic Neural Network

### Import Dependencies

```python
# Import our dependencies
import pandas as pd
import matplotlib as plt
from sklearn.datasets import make_blobs
import sklearn as skl
import tensorflow as tf
```

### Generate datset and graph data

```python
# Generate dummy dataset
X, y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=78)
```

```python
# Creating a DataFrame with the dummy data
df = pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"] = y
print(f"Labels: {y[:10]}")
print(f"Data: {X[:10]}")
```

```python
# Plotting the dummy data
df.plot.scatter(x="Feature 1", y="Feature 2", c="Target", colormap="winter")
```

### Split Data into Testing and Training

```python
# Split dataset into testing and training sets.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

## Normalize dataset 

```python
# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()
```

```python
# Fit the scaler
X_scaler.fit(X_train)
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Create neural network model: Sequential mode

```python
# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()
```

```python
# Add our first Dense layer, including the input layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="relu", input_dim=2))
```

```python
# Add the output layer that uses a probability activation function
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Check the structure of the Sequential model
nn_model.summary()
```

### Train the model

```python
# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

### Visualize model's loss over 100 epochs

```python
# Create a DataFrame containing training history
history_df = pd.DataFrame(fit_model.history, index=range(1,len(fit_model.history["loss"])+1))
```

```python
# Plot the loss
history_df.plot(y="loss")
```

```python
# Plot the accuracy
history_df.plot(y="acc")
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

### Predict new dataset 

```python
# Predict the classification of a new set of blob data
new_X, new_Y = make_blobs(n_samples=10, centers=2, n_features=2, random_state=78)

new_X_scaled = X_scaler.transform(new_X)

nn_model.predict_classes(new_X_scaled)
```
